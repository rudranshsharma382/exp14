```python
import pandas as pd
import numpy as np

data = {
'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
'Price':[55000,20000,30000,2000,8000,45000],
'Units_Sold':[25,60,35,80,50,15]
}

df = pd.DataFrame(data)
print(df)
```

```python
#data binning

bins1=[0,10000,30000,60000]
labels1=['Low','Medium','High']
df['Price_Category']=pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)
```

```python
#Bining units sold
bins2=[0,30,60,100]
labels2=['Low sales','Medium sales','High sales']
df['Units_Category']=pd.cut(df['Units_Sold'],bins=bins2,labels=labels2)

print(df)
```

```python
# Data formatng
# Convert Data type

print(df.dtypes)

df['Units_Sold']=df['Units_Sold'].astype(float)
print(df.dtypes)
print()
print("updates table:")
print(df)
```

```python
#Converting text to Uppercase
df['Product'] = df['Product'].str.upper()


print(df)
```

```python
#Rounding Values
df['Price'] = df['Price'].round(2)
print(df)
```

```python
#Additional Operations
#Sort data
sorted_df = df.sort_values(by='Price',ascending=False) #in higher to lower order of price
print(sorted_df)
print()

sorted_df = df.sort_values(by='Price')#in lower to higher order
print(sorted_df)

```

```python
#Display Unique Values
print(df['Price_Category'].unique())
```

```python
#Dataset:Food Delivery Orders
import pandas as pd
import numpy as np

data = {
'Order_ID':['O1','O2','O3','O4','O5','O6','O7','O8'],
'Order_Value':[250,450,120,600,300,150,500,220],
'Delivery_Time':[30,45,25,60,35,20,50,28],
'Distance_km':[5,8,3,10,6,2,9,4]
}

df = pd.DataFrame(data)

print(df)
```

```python
#Binning Order Value
bins=[0,200,400,700]
labels=['Low','Medium','High']
df['Order_Category']=pd.cut(df['Order_Value'],bins=bins,labels=labels)

print(df)
```

```python
#Binning dilivery time
bins3=[0,25,45,65]
labels3=['Fast','Medium','Slow']
df['Delivery_Category']=pd.cut(df['Delivery_Time'],bins=bins3,labels=labels3)

print(df)
```

```python
#Binning distance
bins4=[0,3,7,11]
labels4=['near','medium','far']
df['Distance_Category']=pd.cut(df['Distance_km'],bins=bins4,labels=labels4)

print(df)
```

```python
#counting the value of categories
print(df['Order_Category'].value_counts())
print()
print(df['Delivery_Category'].value_counts())
print()
print(df['Distance_Category'].value_counts())
```

```python
#Data formatng
# Convert Data type

print(df.dtypes)

df['Delivery_Time']=df['Delivery_Time'].astype(float)
print(df.dtypes)
print()
print("updates table:")
print(df)
```
